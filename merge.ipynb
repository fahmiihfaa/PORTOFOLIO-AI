```python
import pandas as pd

df1 = pd.DataFrame({'id_siswa': ['A001', 'A002', 'A003', 'A004'],
                            'Kelas': ['12A', '12B', '12C', '12D'],
                            'Nilai': [100, 98, 96, 95 ],
                            'Peringkat': [1, 2, 3, 4]},
                             index=[0, 1, 2, 3])

df2 = pd.DataFrame({'id_siswa': ['A005', 'A006', 'A007', 'A008'],
                            'Kelas': ['12E', '12F', '12G', '12H'],
                            'Nilai': [90, 88, 86, 85 ],
                            'Peringkat': [5, 6, 7, 8]},
                             index=[4, 5, 6, 7])

df3 = pd.DataFrame({'id_siswa': ['A009', 'A010', 'A011', 'A012'],
                            'Kelas': ['12I', '12J', '12K', '12L'],
                            'Nilai': [80, 75, 73, 70 ],
                            'Peringkat': [9, 10, 11, 12]},
                            index=[8,9,10,11])
df1 #hanya menampilkan df1
```

```python
list_dataframe = [df1, df2, df3]
pd.concat(list_dataframe) #untuk menggabungkan df1,df2,df3
```

```python
list_dataframe = [df1, df2, df3]
pd.concat(list_dataframe, axis=1) #untuk menggabungkan tetapi jadi kesamping bukan kebawah
```

```python
#merge
#Inner join, left join, right join, full outer join(full join)

df4 = pd.DataFrame({'id_siswa': ['A001', 'A002', 'A003', 'A004'],
                    'Kelas': ['12A', '12B', '12C', '12D'],
                    'Tinggi siswa (cm)': [155, 160, 170, 165 ]},
                     index=[0, 1, 2, 3])
df5 = pd.DataFrame({'id_siswa': ['A001', 'A002', 'A005', 'A006'],
                    'Nilai': [100, 98, 96, 95 ],
                    'Peringkat': [1, 2, 3, 4]},
                     index=[4,5,6,7]) #merge
```

```python
#inner join : data yang sama yang akan dimunculkan
pd.merge(df4, df5, on='id_siswa', how='inner') #dua kolom dari id_siswa lakukan dengan inner join
```

```python
pd.merge(df4, df5, on='id_siswa', how='left') #left dengan df4 ada di left dalam kolom id_siswa dengan left join
```

```python
pd.merge(df4, df5, on='id_siswa', how='outer') #dua kolom dari df4 dan df5 akan muncul semua
```
