```python

import pandas as pd
medan = pd.Series({'beras' :10000, 'telur' :21000, 'cabai' :40000, 'ayam' :35000})
jakarta = pd.Series({'beras' :9000, 'telur' :27000, 'cabai' :45000, 'ayam' :30000})
surabaya = pd.Series({'beras' :11000, 'telur' :24000, 'cabai' :42000, 'ayam' :45000})

df = pd.DataFrame({'medan':medan, 'jakarta':jakarta, 'surabaya':surabaya})
df #membuat tabel (data induk) menggunakan Series
```

```python
#iloc = index locations (melakukan pencarian data berdasarkan lokasi index)
#df.iloc[index_row_start:total_baris , index_column_start:total_column]

df.iloc[1:3, 1:2]
```

```python
#loc : mencari data berdasarkan locasi object
df.loc['telur':'ayam', 'medan':'jakarta']
```

```python
df.mean(axis=1) #menghitung rata-rata baris
```

```python
df.mean() #menghitung rata-rata kolom
```

```python
df['rata-rata'] = df.mean(axis=1) #menambah kolom
df
```

```python
df['total'] = df.sum(axis=1) #menambah kolom jumlah
df
```

```python
df = df.drop('surabaya', axis=1) #hanya bisa untuk menghapus data induk
df
```
